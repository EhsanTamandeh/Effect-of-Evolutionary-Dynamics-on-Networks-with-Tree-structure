```python
import numpy as np 
import random
import networkx as nx
import matplotlib.pyplot as plt
import time 
```

```python
treer7_fix=[]
for i in range (1000):
    treer7_fix.append(np.loadtxt('fixtime_r7_'+str(i)+'.dat'))
```

```python
f7_mean=[]
for f in treer7_fix:
    f7_mean.append(np.sum(f)/len(f))
```

```python
plt.plot(f7_mean,'r')
plt.xlabel('number of added edges to graph');plt.ylabel('fixation time_r7')
plt.ylim(0,350000)
```

```python
np.savetxt('f4.dat',f7_mean)
```

```python
from numpy import arange
from scipy.optimize import curve_fit
plt.loglog(f7_mean,'r')
popt,pcov=curve_fit(lambda x,c,a:c*x**(a),np.arange(0,1000)[5:],f7_mean[5:])
plt.plot(np.arange(0,1000)[5:],popt[0]*np.arange(0,1000)[5:]**(popt[1]))
plt.ylim(0,1000000)
plt.xlim(0,10000)
plt.xlabel('Log number of added edges to graph');plt.ylabel('Log fixation time r=7')
plt.legend()
plt.show()
```

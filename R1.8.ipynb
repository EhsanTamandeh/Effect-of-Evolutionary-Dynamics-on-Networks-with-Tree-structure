```python
import numpy as np 
import random
import networkx as nx
import matplotlib.pyplot as plt
import time 
```

```python
f1=np.loadtxt('fmean0-250.txt')
f2=np.loadtxt('fmean250-500.txt')
f3=np.loadtxt('fmean500-750.txt')
f4=np.loadtxt('fmean750-1000.txt')
```

```python
F_mean=np.zeros(1000)
```

```python
F_mean[0:250]=f1
F_mean[250:500]=f2
F_mean[500:750]=f3
F_mean[750:1000]=f4
```

```python
plt.plot(F_mean)
plt.xlabel('number of added edges to graph');plt.ylabel('fixation time_r1.8')
plt.ylim(0,350000)
```

```python
np.savetxt('f1.dat',F_mean)
```

```python
from numpy import arange
from scipy.optimize import curve_fit
plt.loglog(F_mean,'o-')
popt,pcov=curve_fit(lambda x,c,a:c*x**(a),np.arange(0,1000)[5:],F_mean[5:])
plt.plot(np.arange(0,1000)[5:],popt[0]*np.arange(0,1000)[5:]**(popt[1]))
plt.ylim(0,1000000)
plt.xlim(0,10000)
plt.xlabel('Log number of added edges to graph');plt.ylabel('Log fixation time r=1.8')
plt.legend()
plt.show()
```

```python
lastm_avg=[]
for i in range (1000):
    M_avg=np.loadtxt('M_avg'+str(i)+'.dat')
    lastm_avg.append(M_avg[-1])
```

```python
import numpy as np 
import random
import networkx as nx
import matplotlib.pyplot as plt
import time 
```

```python
treer5_fix=[]
for i in range (1000):
    treer5_fix.append(np.loadtxt('fixtime_r5_'+str(i)+'.dat'))
```

```python
f5_mean=[]
for f in treer5_fix:
    f5_mean.append(np.sum(f)/len(f))
```

```python
plt.plot(f5_mean,'g')
plt.xlabel('number of added edges to graph');plt.ylabel('fixation time_r5')
plt.ylim(0,350000)
```

```python
np.savetxt('f3.dat',f5_mean)
```

```python
from numpy import arange
from scipy.optimize import curve_fit
plt.loglog(f5_mean,'g')
popt,pcov=curve_fit(lambda x,c,a:c*x**(a),np.arange(0,1000)[5:],f5_mean[5:])
plt.plot(np.arange(0,1000)[5:],popt[0]*np.arange(0,1000)[5:]**(popt[1]))
plt.ylim(0,1000000)
plt.xlim(0,10000)
plt.xlabel('Log number of added edges to graph');plt.ylabel('Log fixation time r=5')
plt.legend()
plt.show()
```
